```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("cardio_train.csv",sep=";")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df["age"] = round(df["age"]/365)
df
```

```python
df['BMI'] = df['weight']/(df['height']/100)**2
```

```python
df["BMI_category"] = pd.cut(
    df["BMI"],
    bins=[0, 18.5, 25, 30, 100],
    labels=[0, 1, 2, 3]
)
```
